```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


countrydata = pd.read_csv(r"C:\Users\Saurabh\OneDrive - Manipal Academy of Higher Education\Desktop\MRM-Taskphase\Week_3\Country-data.csv",index_col=0)
print(countrydata)
```

```python
features=np.array(countrydata.loc[:,["gdpp","health","income","life_expec","inflation","imports","child_mort","exports","total_fer"]])
features = (features - features.mean()) / features.std(ddof=0)
features
```

```python
covariance = (1/167)*features.T.dot(features)
covariance
```

```python
import seaborn as sns
plt.figure(figsize=(15,15))
sns.heatmap(covariance, annot=True, square=True)
plt.show()
```

```python
evalues, evector = np.linalg.eig(covariance)

print(evector)
```

```python
print(evalues)
```

```python
np.sum(evalues)
```

```python
explained_variance=(evalues / np.sum(evalues))*100
plt.figure(figsize=(10,5))
plt.bar(range(9), explained_variance, alpha=0.6)
plt.ylabel('Percentage of explained variance')
plt.xlabel('Dimensions')
```

```python
explained_variance=(evalues / np.sum(evalues))*100
explained_variance
```

```python
pc1 = features.dot(evector[:,0])
pc2 = features.dot(evector[:,1])
pc=np.concatenate([[pc1],[pc2]])
pc=pc.T
pc.shape
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
for i in range(features.shape[0]):
    plt.scatter(pc1[i],pc2[i],s=20)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.axhline(y=0, color="grey", linestyle="--")
ax.axvline(x=0, color="grey", linestyle="--")
plt.show()
```

```python
n_clusters=3
k=n_clusters
from numpy.linalg import norm
centroids=np.zeros((3,2))
for i in range(0,k):
    centroids[i,:]=[pc1[np.random.randint(167)],pc2[np.random.randint(167)]]
centroids
```

```python
distance = np.zeros((167,3))
for i in range(0,k):
    for j in range(0,167):
        row_norm=norm(pc[j,:]-centroids[i,:])
        distance[j,i]=np.square(row_norm)
distance
```

```python
labels =np.argmin(distance,axis=1)
labels
```

```python
for i in range(k):
    centroids[i,:]=np.mean(pc[labels==i,:],axis=0)
centroids
```

```python
def distortion(fun,label,cen):
    sums=0
    for i in range(k):
        val=norm(pc[label==i,:]-cen[i,:],axis=1)
        val=np.square(val)
        sums+=np.sum(val)
    return (1/167)*sums
    
```
